```python
# Adjusting the font color of country labels to match the chart colors
import matplotlib.pyplot as plt
import numpy as np
# Data for the charts
# Gender data
gender_labels = ['Male', 'Female']
gender_counts = [6, 5]

# Country data
country_labels = ['Nepal', 'India', 'Iraq', 'Brazil', 'UK', 'Colombia', 'Bolivia']
country_counts = [2, 2, 2, 1,1,2,1]

# Chatbot usage data
chatbot_labels = ['Used Chatbot', 'Did Not Use Chatbot']
chatbot_counts = [8, 3]

# Age data
ages = [28, 25, 26, 31, 39, 29, 35, 33, 31, 23, 21]
age_bins = [20, 25, 30, 35, 40]
age_labels = ['20-25', '26-30', '31-35', '36-40']
# Calculate age distribution
age_distribution, _ = np.histogram(ages, bins=age_bins)

country_colors = [
    'Navy', 'DarkOrange', 'green', 'brown', 'purple', 'DarkKhaki', 'Teal'
]

# Plotting the charts again with updated country label colors
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
fig.tight_layout(pad=3.0)

# Gender pie chart
axes[0, 0].pie(gender_counts, labels=gender_labels, autopct=lambda p: f'{p * sum(gender_counts) / 100:.0f}', colors=['lightblue', 'pink'])
axes[0, 0].set_title('Gender Distribution')

# Country pie chart with colored labels
country_patches, texts, autotexts = axes[0, 1].pie(
    country_counts, labels=country_labels, autopct=lambda p: f'{int(p * sum(country_counts) / 100)}', startangle=140, colors=country_colors
)
for text, color in zip(texts, country_colors):
    text.set_color(color)
axes[0, 1].set_title('Country Distribution')

# Chatbot usage chart
axes[1, 0].bar(chatbot_labels, chatbot_counts, color=['green', 'orange'])
axes[1, 0].set_title('Chatbot Usage')
axes[1, 0].set_ylabel('Number of Participants')

# Age distribution bar chart
axes[1, 1].bar(age_labels, age_distribution, color='purple')
axes[1, 1].set_title('Age Distribution')
axes[1, 1].set_ylabel('Number of Participants')

# Save and show the chart
plt.savefig('/mnt/data/demographic_info_charts_updated.png')
plt.show()
```
